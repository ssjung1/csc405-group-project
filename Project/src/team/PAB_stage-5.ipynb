```python

def col(df, cases):
    if cases:
        c = 'Cases'
    else:
        c = 'Deaths'
    df.rename(columns = {'variable':'Dates', 'value': c, 'First':'First'}, inplace = True)
    
    return df

def addF(df):
    df = df.melt()
    f = np.arange(0, len(df))
    df['First'] = f
    return df
```

```python

def col_l(df, cases):
    if cases:
        c = 'Cases'
    else:
        c = 'Deaths'
    df.rename(columns = {'variable':'Dates', 'value': c, 'First':'First'}, inplace = True)
    
    return df

def addF_l(df):
    #ids = df.iloc[0:1,0:4]    
    df = pd.DataFrame(df.iloc[:,2:401].sum(axis =0)).T
    #dff = pd.concat([ids,df], axis = 1)
    
    df_2 = df.melt()
   
    f = np.arange(0, len(df_2))
    df_2['First'] = f
    #df_2
    return df_2
```
