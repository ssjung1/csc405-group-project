```python
import pandas as pd
import matplotlib.pyplot as plt

#load super set
data = pd.read_csv("cases_deaths_county_data.csv")
data
```

```python
data = data[data['County Name'] != 'Statewide Unallocated']
NewYork = data[data['State'] == 'NY']
NewYork
```

```python

#get most recent 7 days of data
index = 0
for i in reversed(range(len(NewYork.columns))):
    if(NewYork.columns[i][-5:] == 'cases'):
        index = i
        break

#get the most recent 7 days of cases form the NewYork set
NY_last_week_cases = NewYork[NewYork.columns[index - 6:index + 1]]

for i in range(len(NY_last_week_cases.columns)):
    NY_last_week_cases = NY_last_week_cases.rename(columns={NY_last_week_cases.columns[i]: NY_last_week_cases.columns[i][:-6]})
NY_last_week_cases
```

```python
added = NY_last_week_cases.sum()
added.columns = ['Date', 'Cases']
added
```

```python

added.plot(x='Date', y='Cases')
plt.show()
```

The number of Covid-19 cases suddenly jumped between the 10th and 11th of january by around 25,000 people.
This is most likely due to an inconsistant input of data. 

However, we can still see that the overall trend for number of covid cases in New York is still increasing at a fast rate.
