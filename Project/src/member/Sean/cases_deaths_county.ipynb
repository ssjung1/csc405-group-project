```python
import pandas as pd
cases = pd.read_csv("covid_confirmed_usafacts.csv")
deaths = pd.read_csv("covid_deaths_usafacts.csv")
county = pd.read_csv("covid_county_population_usafacts.csv")

cases= pd.melt(frame = cases, id_vars=['countyFIPS','County Name','State','StateFIPS'], var_name= ['Date'], value_name= "Cases")
cases
```

```python
deaths = pd.melt(frame = deaths, id_vars=['countyFIPS', 'County Name', 'State', 'StateFIPS'], var_name=['Date'], value_name="Deaths")
deaths
```

```python
county = county.drop('County Name', axis=1)
county
```

```python
result = pd.merge(cases, deaths, on=['countyFIPS','County Name','State','StateFIPS','Date'])
result
```

```python
SuperSet = pd.merge(result,county)
SuperSet
```

```python
result.to_csv('cases_deaths_county_data.csv')
```
