For one perticular state selection I took california, which has state code "CA"

```python
Cali = Super_covid[Super_covid['State'] == "CA"] 
#apply filter on super dataset for only CA state data and store in Cali variable.  
Cali
```

```python
Cali =Cali[Cali['County Name'] != 'Statewide Unallocated']
# removing all county which named as Statewide Unallocated
Cali
```

```python
Cali['Date'] # finding date range for california state 
```

```python
Cali_Case = Cali.groupby('Date')['Cases'].tail()
#apply groupby funtion on date and cases to count all case and death number
Cali_Case
```

```python
Cali_Case = Cali_Case.tail(7) #here we get last week data by tail()
Cali_Case
```

For plot cases in California in last week 

```python
fig1 = Cali_Case.plot(x='Date',y='Cases',kind = 'line', linestyle='dashed',title='coVid-19 Cases in CALIFORNIA last week')
fig1.set_ylabel("coVid-19 Case")

fig1.legend("Case")
```
