```python
#I chose the quadpoint states Arizona, Colorado, New Mexico, and Utah, with Arizona as the main state.

Arizona = super[super.State == "AZ"]
AZdates = Arizona.drop(['countyFIPS', 'County Name', 'State', 'StateFIPS', 'population'], axis = 1).sum()
#print(AZdates)

AZ_Cases = []
AZ_Deaths = []
for i in range(len(AZdates.keys())):
    if str(AZdates.keys()[i][11:13]) != "2d":
        AZ_Cases.append(AZdates[i])
    else:
        AZ_Deaths.append(AZdates[i])

#print(AZ_Cases)
#print(AZ_Deaths)

#Getting the new cases and deaths
NewAZ_Cases = []
for i in reversed(range(len(AZ_Cases))):
    ANC = AZ_Cases[i] - AZ_Cases[i - 1]
    NewAZ_Cases.insert(0, ANC)

NewAZ_Deaths = []
for i in reversed(range(len(AZ_Deaths))):
    AND = AZ_Deaths[i] - AZ_Deaths[i - 1]
    NewAZ_Deaths.insert(0, AND)

NewAZ_Cases.pop(0)
print(NewAZ_Cases)
NewAZ_Deaths.pop(0)
print(NewAZ_Deaths)
```

```python
#Generating weekly statistics for Arizona

AZC_weeklymeans = []
AZC_weeklymedians = []
AZC_weeklymodes = []
AZD_weeklymeans = []
AZD_weeklymedians = []
AZD_weeklymodes = []

AZC_weekly = []
AZD_weekly = []

for i in range(0, len(NewAZ_Cases), 7):
    divide = slice(i, i + 7, 1)
    AZC_weekly.append(NewAZ_Cases[divide])
print(AZC_weekly)

for i in range(0, len(NewAZ_Deaths), 7):
    divide = slice(i, i + 7, 1)
    AZD_weekly.append(NewAZ_Deaths[divide])
print(AZD_weekly)
```

```python
AZCa = []
AZCb = []
for i in AZC_weekly:
    AZC_weeklymeans = np.mean(i)
    AZCa.append(AZC_weeklymeans)

for i in AZC_weekly:
    AZC_weeklymedians = np.median(i)
    AZCb.append(AZC_weeklymedians)

for i in AZC_weekly:
    AZC_weeklymodes.append(mode(i))

print(AZCa)
print(AZCb)
print(AZC_weeklymodes)

print("")

# AZDc = []
# AZDd = []
for i in AZD_weekly:
    AZD_weeklymeans.append(np.mean(i))
    #AZDc.append(AZC_weeklymeans)

for i in AZC_weekly:
    AZD_weeklymedians.append(np.median(i))
    #AZDd.append(AZC_weeklymedians)

for i in AZC_weekly:
    AZD_weeklymodes.append(mode(i))

print(AZD_weeklymeans)
print(AZD_weeklymedians)
print(AZD_weeklymodes)
```
